# Data Augmentation

```python

import tensorflow as tf
from numpy import expand_dims
from matplotlib import pyplot
from tensorflow import keras
```

## example of vertical shift image augmentation

```python
# load the image
img = keras.preprocessing.image.load_img('images/test6.jpg')
# convert to numpy array
data = keras.preprocessing.image.img_to_array(img)
# expand dimension to one sample
samples = expand_dims(data, 0)
# create image data augmentation generator
datagen = keras.preprocessing.image.ImageDataGenerator(height_shift_range=0.5)
# prepare iterator
it = datagen.flow(samples, batch_size=1)
# generate samples and plot
for i in range(9):
	# define subplot
	pyplot.subplot(330 + 1 + i)
	# generate batch of images
	batch = it.next()
	# convert to unsigned integers for viewing
	image = batch[0].astype('uint8')
	# plot raw pixel data
	pyplot.imshow(image)
	keras.preprocessing.image.save_img('testAugSet6'+str(i)+'.jpg',image)
# show the figure
pyplot.show()
```

## example of rotating image augmentation

```python
# load the image
img = keras.preprocessing.image.load_img('images/test3.png')
# convert to numpy array
data = keras.preprocessing.image.img_to_array(img)
# expand dimension to one sample
samples = expand_dims(data, 0)
# create image data augmentation generator
datagen = keras.preprocessing.image.ImageDataGenerator(rotation_range=90)
# prepare iterator
it = datagen.flow(samples, batch_size=1)
# generate samples and plot
for i in range(9):
	# define subplot
	pyplot.subplot(330 + 1 + i)
	# generate batch of images
	batch = it.next()
	# convert to unsigned integers for viewing
	image = batch[0].astype('uint8')
	# plot raw pixel data
	pyplot.imshow(image)
	keras.preprocessing.image.save_img('testAugSet3_1'+str(i)+'.jpg',image)
# show the figure
pyplot.show()
```
